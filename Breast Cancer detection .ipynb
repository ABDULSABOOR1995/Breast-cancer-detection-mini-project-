### Question 1 (2pts)
Write a function that retrieves the Breast Cancer Wisconsin dataset from Sci-kit Learn's datasets module:
    * https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html
    
Your function should return a dataframe containing the features (n = 30) and target variables from this dataset as shown below:
    * Your dataframe should include 569 rows and 31 columns.

Hint: Use the dataset's DESCR attribute to learn more about the dataset and its features.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer

def cancer_dataset():
    cancer = load_breast_cancer()
    data_frame = pd.DataFrame(np.c_[cancer['data'], cancer['target']],
                  columns= np.append(cancer['feature_names'], ['target']))
    return data_frame
```

```python
df = cancer_dataset()
df
```

```python
def features_and_target(df):
    
    features = df.copy()
    features.drop('target', axis = 1, inplace = True)
    
    target = df['target']
    
    return(features, target)
  
```

### Question 5 (8pts)

Write a function that fits these four classifiers using a percentage-split approach. You will first need to split the dataset into training and testing sets. You should use an iteration statement to apply these classifiers. Use random_state=3000 when splitting your dataset.

```python
# train_test_split 
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.15, random_state=3000)

print("number of training samples:",x_train.shape[0])
print("number of test samples :", x_test.shape[0])
```

```python
def classifiers_percentage_split():
    
    for key, value in estimators.items():
        
        a = value.fit(x_train,y_train)
        yhat = a.predict(x_test)
        accuracy = metrics.accuracy_score(y_test, yhat)
        accuracy = accuracy * 100
        print(key+": ")
        print("         Prediction accuracy on the test data: "+str(round(accuracy, 2))+"% ")
        print("\n")

        
        
    
    
```

```python
classifiers_percentage_split()
```

### Question 6 (8pts)

Write a function that fits these four classifiers using a cross-validation approach. You should use an iteration statement to apply these classifiers. Use random_state=3000 when splitting your dataset.

```python
def classifiers_cross_validation():
    
    for key, value in estimators.items():
        
        a = value.fit(x_train,y_train)
        Rcross = cross_val_score(a, features, target, cv=4)
        mean_accuracy = Rcross.mean()
        mean_accuracy = mean_accuracy * 100
        std = Rcross.std()
        std = std * 100
        print(key+": ")
        print("     mean accuracy = "+str(round(mean_accuracy,2))+"%, standard deviation = "+str(round(std,2))+"%")
        print("\n")

    
```

```python
classifiers_cross_validation()
```

### Best Training/ Testing approach:

There are 2 types of training and testing approaches we have performed in this project. They are <b> percentage-split approach</b> and <b> cross-validation approach</b>.
<b>Cross-validation</b> is a statistical method used to estimate the skill of machine learning models. It gives generalized accuracy of the model. So it is used when we estimating the generalized performance(accuracy) of the model.<br>
Where as <b>percentage-split approach</b> is used when we making the final model. We The training set contains a known output and the model learns on this data in order to be generalized to other data later on. We have the test dataset (or subset) in order to test our model's prediction on this subset. 
In short, we use percentage-split approach when we finalizing our model.
